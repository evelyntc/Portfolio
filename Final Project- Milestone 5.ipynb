```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#connecting to database
sql_connect = sqlite3.connect("Final_Project.db")

#executing SQL query from db
cursor = sql_connect.cursor()
```

```python
#Working with India Pollution table 

#making a variable for my first table India Pollution
india_poll_query = 'SELECT * FROM India_Pollution'

#making a variable for query
results = cursor.execute(india_poll_query).fetchall()

#reading results into a pandas df
india_poll_table = pd.read_sql_query(india_poll_query, sql_connect)
india_poll_table
```

```python
# Working with Polluted Cities table

#making a variable for my second table Polluted cities
poll_cities_query = 'SELECT Country, Year, PM10 FROM Polluted_Cities'

#making a variable for query
results = cursor.execute(poll_cities_query).fetchall()

#reading results into a pandas df
poll_cities_table = pd.read_sql_query(poll_cities_query, sql_connect)
poll_cities_table
```

```python
# Working with US Pollution from 2008-2010 table

#making a variable for my second table Polluted cities
us_poll_query = 'SELECT * FROM US_Pollution'

#making a variable for query
results = cursor.execute(us_poll_query).fetchall()

#reading results into a pandas df
us_poll_table = pd.read_sql_query(us_poll_query, sql_connect)
us_poll_table
```

```python
# India Pollution graphs
```

```python
# Creating a histogram of the SO2 value in Delhi 
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

india_poll_table.hist(ax=ax)
fig.suptitle("SO2 values in Delhi India on March 31st")

#Value of SO2 in Delhi India on March 31st because of Covid-19 lockdowns
```

```python
#Creating a density plot of the SO2 value in Delhi India

sns.distplot(india_poll_table['value'],
            bins=7,
            kde=True,
            vertical=False)
```

```python
# Working with Polluted Cities table
```

```python
# Create a Scatterplot of most polluted cities by PM10

fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=poll_cities_table, 
                x='Country',
                y='PM10',
                sizes=(10,150))
```

```python
#plotting the frequency of the highest PM10
poll_cities_table['PM10'].plot(kind='hist', legend=True)
```

```python
#Grouping Countries with PM10 
poll_cities_group = poll_cities_table.groupby(['Country'])[['PM10']].agg(np.mean)

poll_cities_group
```

```python
poll_cities_group.plot.bar(figsize=(10,7));
```

```python
#Grouping all state NO2 AQI and O3 AQI
us_poll_group = us_poll_table.groupby(['State'])[['NO2 AQI', 'O3 AQI']].agg(np.mean)

us_poll_group
```

```python
us_poll_group.plot.bar(figsize=(30,10));
```
